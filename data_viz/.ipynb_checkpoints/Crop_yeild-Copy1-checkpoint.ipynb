````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
````

````python
crop_yield = pd.read_csv('data/key_crop_yields.csv')
# geo_info = pd.read_json('data/geo-countries.geojson')
crop_yield.head()
````

````python
crop_yield.info()
````

````python
crop_yield.columns
````

````python
#MElt table to create a long table
crop_yield_melt = pd.melt(crop_yield, id_vars=['Entity','Code','Year'], value_vars=['Wheat (tonnes per hectare)',
       'Rice (tonnes per hectare)', 'Maize (tonnes per hectare)',
       'Soybeans (tonnes per hectare)', 'Potatoes (tonnes per hectare)',
       'Beans (tonnes per hectare)', 'Peas (tonnes per hectare)',
       'Cassava (tonnes per hectare)', 'Barley (tonnes per hectare)',
       'Cocoa beans (tonnes per hectare)', 'Bananas (tonnes per hectare)'],
        var_name='crops', value_name='tonnes per hectare')
````

````python
crop_yield_melt.head()
````

````python
# removing unwanted text from column using split
crop_yield_melt['crops'] = crop_yield_melt['crops'].apply(lambda x: x.split(' ')[0])
crop_yield_melt.head()
````

````python
#cleaning column names
crop_yield_melt.columns =crop_yield_melt.columns.str.strip().str.lower().str.replace(' ','_')
#dropping nans
crop_yield_melt =crop_yield_melt[crop_yield_melt['tonnes_per_hectare'].notna()]
crop_yield_melt.info()
````

````python
crop_yield_sort=crop_yield_melt.sort_values(by ='tonnes_per_hectare', ascending =False)
plt.figure(figsize=(10,7))
sns.barplot(x= 'crops', y='tonnes_per_hectare', data =crop_yield_sort, ci=None) # to remove error bar in seaborn use ci(confidence interval =None)
plt.xlabel("")
plt.ylabel("Tonnes Per Hectre",fontsize =15)
plt.xticks(
    rotation= 90, 
    horizontalalignment='center',
    fontweight='light',
    fontsize='x-large')
plt.tight_layout();
````

````python
plt.figure(figsize=(12,9))
sns.lineplot(data=crop_yield_sort, x="year", y="tonnes_per_hectare", hue="crops")
plt.xlabel("")
plt.ylabel("Tonnes Per Hectre",fontsize =15)
plt.xticks(
    rotation= 90, 
    horizontalalignment='center',
    fontweight='light',
    fontsize='x-large')
plt.tight_layout();
````

# Grouped to get maximum production

````python
# crop_yield_melt[['entity','code','year','crops','tonnes_per_hectare']].groupby(['entity','code','year','crops']).agg('sum').sort_values(
#     by ='year', ascending = False)
````

````python
 crop_yield_melt.loc[crop_yield_melt.reset_index().groupby(['crops', 'year'])['tonnes_per_hectare'].idxmax()]
````

````python
# groping by multiple column and getting maximum value within the group
grouped = crop_yield_melt.loc[crop_yield_melt.groupby(['crops', 'year'], sort=False)['tonnes_per_hectare'].idxmax()][['entity','code','year','crops','tonnes_per_hectare']]
grouped = grouped.sort_values(['year','tonnes_per_hectare'],ascending=[True, False]).rename(columns ={'entity':'Country','code':'Code','year':'Year','crops':'Crops','tonnes_per_hectare':'Tonnes per hectare'}).reset_index(drop=True)
grouped["Tonnes per hectare"] = grouped["Tonnes per hectare"].apply(lambda x: round(x, 2))
grouped
# df.sort(['a', 'b'], ascending=[True, False])
````

````python
#checking for year 67
year_67 = grouped[grouped['Year']==1967]
year_67
````

````python
import plotly.express as px

fig_g = px.choropleth(grouped,                            # Input Dataframe
                     locations="Code",           # identify country code column
                     color="Crops",                     # identify representing column
                     hover_data=["Tonnes per hectare", "Country"],             # identify hover name
                     animation_frame="Year",        # identify date column
                     
#                      animation_group ="Country",
                     projection="natural earth",        # select projection
#                      color_continuous_scale = 'Peach',  # select prefer color scale
                     #range_color=[0,10],            # select range of dataset
                    
                     title ="Highest Crop Producing Countries From 1961 - 2018"
                     )  
fig_g.update_layout(
#     font_family="Courier New",
#     font_color = "red"
    font_size= 18,
#     title_font_family="Times New Roman",
#     title_font_color="red",
    title_font_size = 25
#     legend_title_font_color="green"
)
fig_g.show()          
fig_g.write_html("data/crop_yield_map_2.html")                      # write the result to HTML file```B CT
````
